```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, StandardScaler
from google.colab import files
```

```python
uploaded_file_1 = files.upload()
```

```python
courses = list(uploaded_file_1.keys())[0]
df = pd.read_csv(courses)
```

```python
# Tampilkan informasi dasar tentang data
print(df.info())

# Statistik deskriptif
print(df.describe())
```

```python
# Korelasi antara variabel numerik
correlation_matrix = df.corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('courses')
plt.show()
```

```python
# Visualisasi distribusi data
plt.figure(figsize=(10, 6))
sns.histplot(df['code_module'], bins=20, kde=True)
plt.title('code_module')
plt.xlabel('code_presentation')
plt.ylabel('module_presentation_length')
plt.show()
```

```python
# Visualisasi data kategorikal
plt.figure(figsize=(10, 6))
sns.countplot(x='code_presentation', data=df)
plt.title('code_module')
plt.xticks(rotation=45)
plt.show()
```

Preprocessing

```python
df['module_presentation_length'].fillna(df['module_presentation_length'].mean(), inplace=True)
```

```python
df['module_presentation_length'].fillna(df['module_presentation_length'].mode()[0], inplace=True)
```

```python
label_encoder = LabelEncoder()
df['module_presentation_length'] = label_encoder.fit_transform(df['module_presentation_length'])
```
